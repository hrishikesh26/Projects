```python
def str_to_list(rating_str):
    return ast.literal_eval(rating_str)
```

```python
ted.ratings.apply(str_to_list)
```

```python
ted['ratingss_list'] = ted.ratings.apply(lambda x: ast.literal_eval(x))
```

```python
ted.ratingss_list
```

```python
#def get_num(list_of_dicts):
#    return list_of_dicts[0] 
```

```python
def get_num(list_of_dicts):
    num=0
    for d in list_of_dicts:
        num = num + d['count']
    return num
```

```python
get_num(ted.ratingss_list[0])
```

```python
ted.ratingss_list.apply(get_num)
```

```python
ted['num_ratings']=ted.ratingss_list.apply(get_num)
```

```python
ted.num_ratings.describe()
```

# which occupations delivered the funniest TED talks on average?

### step1: Count the number of funny ratings

```python
ted.ratingss_list.head()
```

```python
ted.ratings.str.contains('Funny').value_counts()
```

```python
def get_funnny_ratings(list_of_dicts):
    for d in list_of_dicts:
        if d['name']=='Funny':
            return d['count']
```

```python
ted.ratingss_list.apply(get_funnny_ratings)
```

```python
ted['funny_ratings']=ted.ratingss_list.apply(get_funnny_ratings)
```

```python
ted.sort_values('funny_ratings').speaker_occupation.tail(10)
```
